```python
def irradiated_isomer_selector (irradiated_isomer_code):
    global node_from_long_code
    matched_isomer_codes = []
    for df_record in table_of_df: #df bound to the dictionaries we have
        if  df_record['irradiated_isomer_code'] == irradiated_isomer_code:
             matched_isomer_codes.append(df_record)
    if not (len(matched_isomer_codes)>2):
        warnings.warn('more than 2 entries found for isomer code '+ irradiated_isomer_code)
    elif not (len(matched_isomer_codes)==0):
        warnings.warn('zero entries found for isomer code '+ irradiated_isomer_code)
    
    currently_selected_isomer={'experimental': None,'fitted': None, 'irradiated_isomer_code': irradiated_isomer_code, 'substituent_type': 'phenyl', 'short_code': node_from_long_code[irradiated_isomer_code]['label_label']}
    for matched_isomer in matched_isomer_codes:
        if matched_isomer ['exp_or_fitted']=='experimental':
            currently_selected_isomer.update({'experimental':matched_isomer})
        elif matched_isomer ['exp_or_fitted']=='fitted':
            currently_selected_isomer.update({'fitted':matched_isomer})
        else:
            warnings.warn('neither fitted nor experimental: '+ matched_isomer ['exp_or_fitted'])
    return currently_selected_isomer

```
